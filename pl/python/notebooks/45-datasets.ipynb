## PyDataset

```python
# PyDataset

from pydataset import data

# όλα τα διαθέσιμα datasets
data()
```

```python
# Πληροφορίες για το dataset titanic
data('Housing', show_doc=True)
```

```python

df = data('Housing')

# αλλαγή δεικτών έτσι ώστε να ξεκινούν από 0 αντί για 1
# import numpy as np
# df.index = np.arange(0, len(df))

df.head()
```

```python
# Πληροφορίες για το DataFrame

df.info()
```

```python
# Στατιστικά για το DataFrame

df.describe()
```

```python
# Επιλογή συγκεκριμένων γραμμών, στηλών (με το loc)

df.loc[1:3, ['price', 'bedrooms']]
```

```python
# Επιλογή συγκεκριμένων γραμμών (με το iloc)

df.iloc[1:3]
```

```python
# Δειγματοληψία εγγραφών

df.sample(10)
```

## Statsmodels

```python
import statsmodels.api as sm

df = sm.datasets.longley.load_pandas()['data']
print(df.__doc__)

df
```

## Datasets στο module sklearn

```python
import pandas as pd
from sklearn import datasets

iris = datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

```python
df.sample(10)
```
