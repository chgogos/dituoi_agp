# Introspection

Introspection είναι η δυνατότητα ενός προγράμματος να εξετάζει τη δομή και τα χαρακτηριστικά των αντικειμένων του κατά την εκτέλεση.

Επιτρέπει τον εντοπισμό πληροφοριών όπως:
- τον τύπο ενός αντικειμένου
- τις ιδιότητες και μεθόδους του
- τα ορίσματα μιας συνάρτησης
- τον πηγαίο κώδικα

```python
# 1. Inspecting objects

x = "hello"

type(x)        # <class 'str'>
dir(x)         # list of attributes/methods
```

```python
# 2. Accessing attributes dynamically
class A:
    def __init__(self):
        self.value = 42

obj = A()

getattr(obj, "value")     # 42
setattr(obj, "value", 100)
hasattr(obj, "value")     # True
```

```python
# 3. Calling methods dynamically

class A:
    def greet(self):
        return "hi"

obj = A()

method = getattr(obj, "greet")
method()   # "hi"
```

```python
# 4. Inspecting functions and classes

import inspect

def f(a, b): pass

inspect.signature(f)      # (a, b)
inspect.getsource(f)      # source code
```

```python
# 5. Dynamic class loading (closest to Java-style reflection)

module = __import__("math")
func = getattr(module, "sqrt")
func(16)   # 4.0
```
