# Pillow

```python
# φόρτωση αρχείου jpg

from PIL import Image

img = Image.open(r"../../../images/cleese-silly-walks1.jpg")
display(img)
```

```python
# σμίκρυνση αρχείου εικόνας

original_size = img.size
print(original_size)
new_size = [int(x*0.4) for x in original_size]
print(new_size)
img2 = img.resize(new_size, resample = Image.BILINEAR)
display(img2)
```

```python
# δημιουργία thumbnail εικόνας

img3 = Image.open(r"../../../resources/cleese-silly-walks1.jpg")
img3.thumbnail((30,30))
display(img3)
```

```python
# σχεδίαση μιας εικόνας από το μηδέν
# https://stackoverflow.com/questions/26649716/how-to-show-pil-image-in-ipython-notebook

from PIL import Image

img = Image.new( 'RGB', (255,255), "black") # Create a new black image
pixels = img.load() # Create the pixel map
for i in range(img.size[0]):    # For every pixel:
    for j in range(img.size[1]):
        pixels[i,j] = (i, j, 100) # Set the colour accordingly

display(img)
```
