# Seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
iris_df = pd.read_csv("../../../datasets/iris.csv")
iris_df
```

```python
sns.scatterplot(data=iris_df,x="petal_length", y="petal_width")
```

```python
sns.boxplot(data=iris_df, x="species", y="petal_length")
```

## Αλλαγή εμφάνισης γραφημάτων

```python
# πλέγμα

# sns.set_style("darkgrid") # πλέγμα με σκούρο υπόβαθρο
# sns.set_style("whitegrid") # πλέγμα με λευκό υπόβαθρο
sns.set_style("ticks") # χωρίς πλέγμα
sns.scatterplot(
    data=iris_df, 
    x="petal_length", 
    y="petal_width", 
    hue="species", 
    )
```

```python
# συνδυασμοί χρωμάτων

sns.scatterplot(
    data=iris_df, 
    x="petal_length", 
    y="petal_width", 
    hue="species", 
    # palette="tab10" # default
    # palette="pastel"
    # palette="colorblind"
    # palette="deep"
    # palette="muted"
    palette="dark"
    # palette="bright"
    )
```

```python
# τίτλοι και ετικέτες αξόνων

ax = sns.scatterplot(
    data=iris_df, 
    x="petal_length", 
    y="petal_width", 
    hue="species", 
    palette="dark"
    )

ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
ax.set_title("IRIS Dataset")
```
